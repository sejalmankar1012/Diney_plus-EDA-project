## Filling the null values of 'director' and 'cast' with hard coded string 'unknown' 

```python
null_filled['director'].fillna('Unknown',inplace=True)
null_filled['cast'].fillna('Unknown',inplace=True)
null_filled.isnull().sum()
```

## Filling the null values of 'country' and 'date added' with mode

```python
null_filled['country'].fillna(null_filled['country'].mode()[0],inplace=True)
null_filled['date_added'].fillna(null_filled['country'].mode()[0],inplace=True)
null_filled
```

```python
null_filled['rating'].value_counts()
```

```python
null_filled['rating']=null_filled['rating'].map(null_filled['rating'].value_counts().to_dict())
null_filled
```

## To fill the null values of rating as we dont know the exact ratings we can fill it with mean as we have already converted the ratings in numerical format

```python
null_filled['rating'].fillna(null_filled['rating'].mean(),inplace=True)
```

```python
null_filled.isnull().sum()
```
